```python
# Classify movies based on ratings [excellent, good and average]

data.head()
```

```python
def rating(rat):
    if rat>=7.0:
        return "Excellent"
    elif rat>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['rating_category'] = data["Rating"].apply(rating)
```

```python
#Find unique values from genre

data['Genre']
```

```python
#First step: one common list

list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
#2nd step - 2D list to 1D list

one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
# now we actually find the unique values in genre values:

uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```
